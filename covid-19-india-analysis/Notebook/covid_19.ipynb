### Data Cleaning


```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
df['Date']
```

```python

df['Day'] = df['Date'].dt.day## Create day colunm
df['Month'] = df['Date'].dt.month_name()## Create month colunm
df['Year'] = df['Date'].dt.year## Create year colunm
```

```python
df.head()
```

```python
df = df.drop(columns=['Date'])## drop the date columns because  sepreated day, month& year.
```

```python
df.head()
```

```python
df.info()
```

```python
## Chnage thier data type
df['ConfirmedIndianNational'] = pd.to_numeric( df['ConfirmedIndianNational'], errors='coerce')
df['ConfirmedForeignNational'] = pd.to_numeric(df['ConfirmedForeignNational'], errors='coerce')
df['Month'] = df['Month'].astype('category')
```

```python
df.info()
```

```python
##check missing values
df.isnull().sum()
```

```python
## Handling missing values
## In this  17664 missing values so its 98% data is null so it's better to drop that colunm
df=df.drop(columns=['ConfirmedIndianNational'])
df=df.drop(columns=['ConfirmedForeignNational'])
```

```python
df.isnull().sum()
```

##### The dataset was cleaned by handling date formats and separating the date column into day, month, and year. This made time-based analysis more flexible and interpretable.


### Feature Engineering

```python
## Create New Columns
df['Active'] = df['Confirmed'] - df['Cured'] - df['Deaths']
```

```python
df['Daily_Confirmed'] = df.groupby('State/UnionTerritory')['Confirmed'].diff().fillna(0)
df['Daily_Confirmed'] = df['Daily_Confirmed'].astype(int)## change data type
```

```python
df.info()
```

### State-wise Analysis
##### State-wise aggregation helps identify regions that were most and least affected by COVID-19, providing a clearer picture of regional impact.


```python
## Total COVID Cases in India
total_confirmed = df['Confirmed'].max()
total_deaths = df['Deaths'].max()
total_recovered = df['Cured'].max()

total_confirmed, total_deaths, total_recovered
```

```python
## Top 10 States by Confirmed Cases
top_states = (
    df.groupby('State/UnionTerritory')['Confirmed']
    .max()
    .sort_values(ascending=False)
    .head(10)
)

top_states
```

```python
## Plot
plt.figure(figsize=(10,5))
top_states.plot(kind='bar')
plt.title("Top 10 States by Confirmed COVID-19 Cases")
plt.ylabel("Confirmed Cases")
plt.show()
```

### Time-Series Analysis (India Overall)

#####  Analyzing COVID-19 cases over time helps in identifying waves, peak periods, and overall progression of the pandemic.


```python
## India-level Daily Data
india_daily = df.groupby(['Year','Month','Day'], observed=False)[['Confirmed','Deaths','Cured']].sum().reset_index()
india_daily.head()
```

```python
## Confirmed Cases Trend
plt.figure(figsize=(12,5))
plt.plot(india_daily.index, india_daily['Confirmed'])
plt.title("COVID-19 Confirmed Cases Trend in India")
plt.xlabel("Date")
plt.ylabel("Confirmed Cases")
plt.show()
```

```python
## Daily New Cases Trend

india_daily['Daily_New'] = india_daily['Confirmed'].diff()

plt.figure(figsize=(12,5))
plt.plot(india_daily.index, india_daily['Daily_New'])
plt.title("Daily New COVID-19 Cases in India")
plt.xlabel("Date")
plt.ylabel("Daily Cases")
plt.show()
```

### State-wise Comparison

```python
## Select Major States
states = ['Maharashtra', 'Kerala', 'Delhi', 'Tamil Nadu']
```

```python
plt.figure(figsize=(12,6))

for state in states:
    state_data = df[df['State/UnionTerritory'] == state]
    plt.plot(state_data['Day'], state_data['Confirmed'], label=state)

plt.legend()
plt.title("State-wise COVID-19 Trend")
plt.xlabel("Day")
plt.ylabel("Confirmed Cases")
plt.show()
```

#### Month-wise COVID Case Analysis 

```python
monthly_cases = ( df.groupby(['Year','Month'], observed=True)['Confirmed'].sum().reset_index())
monthly_cases.head()
```

```python
plt.figure(figsize=(10,4))
plt.plot(monthly_cases['Confirmed'])
plt.title("Month-wise COVID-19 Cases in India")
plt.ylabel("Confirmed Cases")
plt.show()

```

#### Top 5 States Comparison Over Time

```python
top_states = df.groupby('State/UnionTerritory')['Confirmed'].max().sort_values(ascending=False).head(5).index

plt.figure(figsize=(10,5))
for state in top_states:
    data = df[df['State/UnionTerritory'] == state]
    plt.plot(data['Confirmed'], label=state)

plt.legend()
plt.title("Top 5 States COVID-19 Trend Comparison")
plt.ylabel("Confirmed Cases")
plt.show()
```

#### Year-wise COVID Growth

```python
yearly_cases = df.groupby('Year')['Confirmed'].sum()

yearly_cases.plot(kind='bar', figsize=(6,4), title='Year-wise COVID-19 Growth')
plt.ylabel("Confirmed Cases")
plt.show()
```

#### When did each state hit its highest cases?

```python
peak_cases = df.groupby('State/UnionTerritory')['Confirmed'].idxmax()
state_peaks = df.loc[peak_cases, ['State/UnionTerritory','Confirmed','Day','Month','Year']]

state_peaks.head()
```

#### Was the second wave worse than the first?

```python
wave1 = df[df['Year'] == 2020]['Confirmed'].sum()
wave2 = df[df['Year'] == 2021]['Confirmed'].sum()

wave1, wave2
```

```python
plt.bar(['First Wave','Second Wave'], [wave1, wave2])
plt.title("COVID-19 First vs Second Wave")
plt.ylabel("Confirmed Cases")
plt.show()
```

#### Which states had the lowest COVID impact?

```python
least_states = (df.groupby('State/UnionTerritory')['Confirmed'].max().sort_values().head(10))
least_states.plot(kind='bar', figsize=(10,5), title='Least Affected States')
plt.ylabel("Confirmed Cases")
plt.show()
```

#### Which state ranked highest each month?

```python
monthly_state = df.groupby(['Year','Month','State/UnionTerritory'],observed=True)['Confirmed'].max().reset_index()

monthly_rank = monthly_state.sort_values(['Year','Month','Confirmed'], ascending=[True, True, False])

monthly_rank.head()
```

#### Worst Month in India

```python
worst_month = (df.groupby(['Year','Month'],observed=True)['Confirmed'].sum().reset_index()
               .sort_values('Confirmed', ascending=False)
               .head(1)
)

worst_month
```

```python
month_dist = df.groupby('Month', observed=True)['Confirmed'].sum()
month_dist.plot(kind='bar', figsize=(8,5), title='Month-wise Highst COVID Case Distribution')
plt.ylabel("Confirmed Cases")
plt.show()
```

#### Which states contributed most overall?

```python
state_contribution = ( df.groupby('State/UnionTerritory')['Confirmed'].max().sort_values(ascending=False))

state_contribution.head(10)
```

##### Which states fluctuated the most?

```python
volatility = df.groupby('State/UnionTerritory')['New_Cases'].std().sort_values(ascending=False)
volatility.head()
```

##### Recovery vs Death Comparison

##### This comparison highlights the severity of COVID-19 by contrasting recovery outcomes with mortality across major states.


```python
comparison = df.groupby('State/UnionTerritory')[['Cured','Deaths']].max()
comparison.head()
```

```python
state_recovery_death = ( df.groupby('State/UnionTerritory')[['Cured','Deaths']].max()
    .sort_values(by='Cured', ascending=False)
    .head(10)   # top 10 states for clarity
)
```

```python
state_recovery_death.plot(kind='bar',figsize=(10,5))

plt.title("Recovery vs Death Comparison (Top 10 States)")
plt.ylabel("Number of Cases")
plt.xlabel("State")
plt.xticks(rotation=45)
plt.show()
```

#### Cumulative National Trend

```python
national_trend = df.groupby(['Year','Month'],observed=True)[['Confirmed','Cured','Deaths']].sum()
national_trend.plot(figsize=(10,5))
plt.title("India COVID-19 Cumulative Trend")
plt.show()
```

##### This analysis explores COVID-19 data in India to understand how cases varied across different states and over time. The findings show that the impact of COVID-19 was not uniform, with Maharashtra being the most affected state. Monthly trends highlighted clear waves of infection during the pandemic.

##### Overall, recovery cases were significantly higher than death cases across most states, indicating a strong recovery trend. This analysis demonstrates how data cleaning, grouping, and visualization techniques can be used to extract meaningful insights from real-world data.
